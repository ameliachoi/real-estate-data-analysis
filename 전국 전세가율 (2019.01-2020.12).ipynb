```python
sell = pd.read_csv('평균단위매매가격_아파트_csv.csv')
jeonse = pd.read_csv('평균단위전세가격_아파트_csv.csv')
```

```python
sell.head(3)
```

```python
jeonse.head(3)
```

```python
# numpy 연산 위해 숫자만 분리
sell_num = sell.iloc[:, 1:]
sell_num.head()
```

```python
jeonse_num = jeonse.iloc[:, 1:]
jeonse_num.head()
```

```python
sell_num_np = sell_num.to_numpy()
jeonse_num_np = jeonse_num.to_numpy()
```

```python
gap_num = (jeonse_num_np / sell_num_np) * 100
gap_num
```

```python
gap = pd.DataFrame(gap_num, columns = sell_num.columns)
gap.head(3)
```

```python
gap = pd.concat([sell.iloc[:, 0], gap], axis=1)
gap.head(3)
```

```python
gap = gap.set_index('지 역')
```

```python
gap.info()
```
